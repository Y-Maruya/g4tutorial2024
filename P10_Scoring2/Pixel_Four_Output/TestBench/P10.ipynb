```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("P10_Pixel_Four_Output.csv", header=None)

#extract data frames according to the Pixel replica number in the column 3
pixel = [df[df.iloc[:, 3] == 1000], df[df.iloc[:, 3] == 1001], df[df.iloc[:, 3] == 1002], df[df.iloc[:, 3] == 1003]]

# mean and sigma of coordinate y annd x at pixel #1~4
for i in range(4):
    mean_value = pixel[i][5].mean()
    sigma = pixel[i][5].std()
    print("Y at Pixel %i , mean = %.4f,  sigma = %.4f" %(i+1, mean_value, sigma))
    mean_value = pixel[i][4].mean()
    sigma = pixel[i][4].std()
    print("X at Pixel %i , mean = %.4f,  sigma = %.4f" %(i+1, mean_value, sigma))

# coordinate y including all pixels
plt.hist(df[5],bins=400)
plt.xlabel(" Y (mm)")
plt.ylabel("Frequency")
plt.title("coord Y at Pixel #1~#4")
plt.savefig("WorldY")
plt.show()
# XY scatter plots for each pixels
for i in range(4):
    plt.scatter(pixel[i][4], pixel[i][5], c="red", s=1, marker = ".")
    plt.xlabel(" X (mm)")
    plt.ylabel(" Y (mm)")
    title = "X vs Y at pixel %i" %(i+1)
    plt.title(title)
    figname = "scatterXY_%i" %(i+1)
    plt.savefig(figname)
    plt.show()
```
